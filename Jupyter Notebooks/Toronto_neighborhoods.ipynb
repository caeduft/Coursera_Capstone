### 1.1 Packages required

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 
```

### 1.2 Request the web page and parse it with BS4

```python
url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")

whole_page = BeautifulSoup(url.content,'html.parser')
# print(whole_page.prettify()) #Prints a really long output, useful to check if the request and parsing were ok

all_tables=whole_page.find_all("table")
len(all_tables)
```

#### 1.2.1 There are other 2 tables in the web page. Let's look for the one of interest.

```python
postcode_html = whole_page.find('table', class_='wikitable sortable') # all_tables[0] would have also worked

postcode_html #Prints a really long output, useful to check if the request and parsing were ok
# pd.read_html(str(postcode_html)) #Requires lxml
```

#### 1.2.2 Start parsing the table. Start with the headers. Doing this, will give an initial indication if the html tags are properly used

```python
headers=[]

for row in postcode_html.findAll('tr'):
    cells=row.findAll('th')
    for i in range(len(cells)):
        headers.append(cells[i].find(text=True))

headers    
```

#### 1.2.3 Remove non printable characters

```python
clean_headers = [h.replace("\n", '') for h in headers]
clean_headers
```

#### 1.2.4 Parse the body of the table. Check for the length of each row as this will reveal split or merged cells (currently it will only ommit those cells)

```python
PC = []
BO = []
NB = []
expected_cols=3
for row in postcode_html.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==expected_cols:
        PC.append(cells[0].find(text=True))
        BO.append(cells[1].find(text=True))
        NB.append(cells[2].find(text=True))
        
print(PC,BO,NB)
```

#### 1.2.5 Remove non printable characters

```python
clean_PC = [h.replace("\n", '') for h in PC]
clean_BO = [h.replace("\n", '') for h in BO]
clean_NB = [h.replace("\n", '') for h in NB]
clean_columns=[clean_PC,clean_BO,clean_NB]
clean_columns
```

### 1.3 Convert the lists into a dataframe

```python
postcode_df=pd.DataFrame(clean_columns).T

postcode_df.columns = clean_headers
postcode_df
```

#### 1.3.1 Remove non assigned boroughs

```python
clean_df = postcode_df[postcode_df["Borough"]!="Not assigned"]
clean_df
```

#### 1.3.2 Check if there are non assigned neighborhoods

```python
(clean_df["Neighbourhood"]=="Not assigned").sum()
```

#### 1.3.3 Check if there are repeated postcodes

```python
len(clean_df["Postal Code"].unique())==clean_df.shape[0]
```

```python
print("The clean dataset contains {} rows".format(clean_df.shape[0]))
```

## 3.3 Request venues

```python
CLIENT_ID = FS_ID["ID"] # your Foursquare ID
CLIENT_SECRET = FS_ID["Secret"] # your Foursquare Secret
VERSION = '20180605' # Foursquare API version
LIMIT = 100 # A default Foursquare API limit value
```

### 3.3.2 Request the venues

#### 3.3.2.1 Test run only with the first borough (North York)

##### 3.3.2.1.1 Define URL parameters

```python
latitude = nb_df.Latitude[0]
longitude = nb_df.Longitude[0]
radius = 500
url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, radius, LIMIT)
```

##### 3.3.2.1.2 Request the venues

```python
results = requests.get(url).json()
results
```

##### 3.3.2.1.3 Define a function to extract categories from the venues

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

##### 3.3.2.1.4 Extract categories from the results

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

#### 3.3.2.2 Request venues for all boroughs

##### 3.3.2.2.1 Define a function to iterate requests and parse through every borough

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

##### 3.3.2.2.2 Request the venues

```python
toronto_venues = getNearbyVenues(nb_df['Neighbourhood'], nb_df['Latitude'], nb_df['Longitude'], radius=500)
toronto_venues.head()
```

### 3.3 Explore the neighbourhoods

#### 3.3.1 One hot encoding of the venues' categories

**Notice there is a category named neighborhood**  This column will be renamed to NB to avoid losing this info

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
toronto_onehot.rename(columns={"Neighborhood": "NB"}, inplace=True)

toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 
toronto_onehot.head()
# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

#### 3.3.2 Look at the normalized frequencies of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
print("There are {1} venue categories for {0} neighborhoods".format(toronto_grouped.shape[0],toronto_grouped.shape[1]-1))
```

#### 3.3.3 Check the 3 top venues for each neighborhood

```python
num_top_venues = 3

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

#### 3.3.4 Place the 10 top venues for each neighborhood in a dataframe

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

print(neighborhoods_venues_sorted.shape)
neighborhoods_venues_sorted.head()
```

### 3.4 Cluster the neighborhoods based on their venue categories

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:96] 
```

#### 3.4.1 Merge the k means labels with the previous dataframes

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

#Remove duplicated neighborhoods (i.e. Downville that spans multiple post codes)
toronto_merged = nb_df.drop_duplicates(subset=['Neighbourhood'],keep="first")

# merge toronto_merged with nb_df to add latitude/longitude for each neighborhood. Make explicit inner to remove neighborhoods without venues
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighbourhood', how="inner")

toronto_merged.head(10) # check the last columns!
```

#### 3.4.2 Visualize the labels

```python
# create map
map_clusters = folium.Map(location=[43.670278, -79.386667], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-1],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
